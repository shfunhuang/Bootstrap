```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python
iris = datasets.load_iris()

x = pd.DataFrame(iris['data'], columns=iris['feature_names'])
#print("target: " + str(iris['target']))
print("target_names: " + str(iris['target_names']))
#print("feature_names: " + str(iris['feature_names']))

y = pd.DataFrame(iris['target'], columns=['target'])
iris_data = pd.concat([x,y], axis=1)
iris_data.head(3)
```

```python
target_name = {
    0:'setosa',
    1:'versicolor',
    2:'virginica'
}
target_name
```

```python
iris_data['target_name'] = iris_data['target'].map(target_name)
iris_data = iris_data[(iris_data['target_name'] == 'setosa')|(iris_data['target_name'] == 'versicolor')]
iris_data = iris_data[['sepal length (cm)','petal length (cm)','target_name']]
iris_data.head(5)
```

```python
target_class = {
    'setosa':1,
    'versicolor':-1
}
target_class
```

```python
iris_data['target_class'] = iris_data['target_name'].map(target_class)
```

```python
# Activation function
def sign(z):
    if z > 0:
        return 1
    else:
        return -1
```

```python
w = np.array([0.,0.,0.])
x,y = np.concatenate((np.array([1.]), np.array(iris_data.iloc[0])[:2])), np.array(iris_data.iloc[0])[2]
print np.array([1.]);
print iris_data.iloc[0];
print x,y
```

```python
#w=[0,0,0] 0*1 + 0* + 0* = 0 0 versicolor(-1)
w = np.array([0.,0.,0.]) 
error = 1
iterator = 0
min_e = np.dot(w,x) #initial minimum error

while error != 0:
    error = 0

    iris_data['randNumCol'] = np.random.choice(iris_data.shape[0], iris_data.shape[0], replace=False)
    iris_data_sample = iris_data.sort_values(by = 'randNumCol')[:1]

    # xx0=1 (x01)
    x = np.concatenate((np.array([1.]), np.array(iris_data_sample.iloc[0])[:2]))
    y = np.array(iris_data_sample.iloc[0])[2]

    wt = w + y*x #review weight
    print("sign: " + str(sign(np.dot(wt,x))) + ", y: " + str(y) + ", x: " + str(x))
    print("dot: " + str(np.dot(wt,x)) + ", min_e: " + str(min_e))

    if(np.dot(wt,x) < min_e):
        w = wt
        min_e = np.dot(wt,x)

    print("iterator: " +str(iterator))
    iterator += 1

    #
    error += 1

    sns.lmplot('sepal length (cm)','petal length (cm)',data=iris_data, fit_reg=False, hue ='target_class')
            
    # Decision boundary 
    if w[1] != 0:
        x_last_decision_boundary = np.linspace(0,w[1])
        y_last_decision_boundary = (w[2]/w[1])*x_last_decision_boundary
        plt.plot(x_last_decision_boundary, y_last_decision_boundary,'c--')

    print("x: " + str(x))
    print("w: " + str(w))


    # x 
    x_vector = np.linspace(0,x[1])
    y_vector = (x[2]/x[1])*x_vector
    plt.plot(x_vector, y_vector,'b')

    # Decision boundary 
    x_decision_boundary = np.linspace(-0.5,7)
    y_decision_boundary = (-w[1]/w[2])*x_decision_boundary - (w[0]/w[2])
    plt.plot(x_decision_boundary, y_decision_boundary,'r')

    # Decision boundary 
    x_decision_boundary_normal_vector = np.linspace(0,w[1])
    y_decision_boundary_normal_vector = (w[2]/w[1])*x_decision_boundary_normal_vector
    plt.plot(x_decision_boundary_normal_vector, y_decision_boundary_normal_vector,'g')
    plt.xlim(-0.5,7.5)
    plt.ylim(5,-3)
    plt.show()

    if(iterator == 10):
        break
```
